```python
import pandas as pd

file_path = "POWER_Point_Monthly_20140101_20241231_013d09N_077d58E_LST.csv"

# =========================
# STEP 1: READ RAW LINES
# =========================

with open(file_path, "r") as f:
    lines = f.readlines()

# =========================
# STEP 2: FIND START OF DATA
# =========================

start = 0
for i, line in enumerate(lines):
    if line.strip().startswith("PARAMETER"):
        start = i
        break

# =========================
# STEP 3: READ CLEAN TABLE
# =========================

weather = pd.read_csv(
    file_path,
    skiprows=start,
    sep=",",
    engine="python"
)

print(weather.head())
print(weather.columns)

# =========================
# STEP 4: CLEAN DATA
# =========================

# Clean column names
weather.columns = weather.columns.str.strip().str.upper()

# Drop ANN
weather = weather.drop(columns=["ANN"], errors="ignore")

# Fill missing PARAMETER values (safe)
weather["PARAMETER"] = weather["PARAMETER"].ffill()

# Keep only rainfall
weather = weather[weather["PARAMETER"] == "PRECTOTCORR"]

# Convert YEAR
weather["YEAR"] = pd.to_numeric(weather["YEAR"], errors="coerce")

# =========================
# STEP 5: MELT
# =========================

weather_long = weather.melt(
    id_vars=["YEAR"],
    var_name="month",
    value_name="rainfall"
)

# 🔥 IMPORTANT FIX (THIS WAS MISSING)
weather_long["month"] = weather_long["month"].astype(str).str.strip().str.upper()

# Month mapping
month_map = {
    "JAN": 1, "FEB": 2, "MAR": 3, "APR": 4,
    "MAY": 5, "JUN": 6, "JUL": 7, "AUG": 8,
    "SEP": 9, "OCT": 10, "NOV": 11, "DEC": 12
}

weather_long["month"] = weather_long["month"].map(month_map)

# Convert rainfall
weather_long["rainfall"] = pd.to_numeric(weather_long["rainfall"], errors="coerce")

# Rename
weather_long.rename(columns={"YEAR": "year"}, inplace=True)

# Drop invalid rows (important)
weather_long = weather_long.dropna(subset=["month", "rainfall"])

print("✅ CLEANED WEATHER:")
print(weather_long.head())

# =========================
# STEP 6: MERGE
# =========================

df = pd.read_csv("final_dataset.csv")

df = pd.merge(df, weather_long, left_on="Crop_Year", right_on="year", how="left")

df.to_csv("final_dataset_with_weather.csv", index=False)

print("✅ DONE - WEATHER MERGED")
```
